```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Creating the enriched dataset
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank', 'Grace', 'Helen', 'Ivy', 'Jack',
             'Alice', 'Eva', 'Charlie', 'Eva'],  # Duplicate names
    'Age': [25, 30, 35, np.nan, 40, 50, -999, 28, None, 45, 25, 30, 35, 40],  # Missing values, None, Outlier (-999)
    'Gender': ['Female', 'Male', 'male', 'Female', 'Female', np.nan, 'Female', 'f', 'Female', 'Male',
               'Female', 'Female', 'Male', 'Female'],  # Inconsistent capitalization
    'Math_Score': [88, 92, 100, 78, 85, np.nan, 110, 95, 60, 88, 88, np.nan, 'N/A', 85],  # Missing values, Outliers, Incorrect data type
    'Science_Score': [np.nan, 85, 95, 70, 'missing', 98, 100, 105, 60, 77, 85, 85, 70, 70],  # Missing values, Inconsistent data representation
    'Enrollment_Date': ['2020-01-15', '15/02/2020', '2020/03/10', 'April 5, 2020', np.nan, '2020-05-25',
                        '2020-06-30', 'July 10, 2020', '2020-08-15', '2020-09-01', '2020-01-15', '2020-05-25',
                        '2020/03/10', 'April 5, 2020'],  # Inconsistent date formats, Missing values
    'Graduated': ['Yes', 'No', 'No', 'Yes', 'No', np.nan, 'yes', 'No', 'Yes', 'no', 'Yes', 'No', 'No', 'Yes'],  # Inconsistent capitalization
}

# Converting into a DataFrame
df = pd.DataFrame(data)
df
```

```python
df[df.duplicated(subset = ['Name','Enrollment_Date'])]
```

```python
df.drop_duplicates(subset =['Name', 'Enrollment_Date'], inplace = True)
df
```

```python
df['Age'].replace(-999,np.nan, inplace = True)
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
df['Science_Score'].replace('missing',np.nan, inplace = True)
df
```

```python
df['Gender'] = df['Gender'].str.capitalize()
df['Gender'].replace(np.nan, 'Male', inplace = True)
```

```python
df['Gender'].replace('F',' Female', inplace = True)
```

```python
df
```

```python
df['Graduated'] = df['Graduated'].str.capitalize()
mode_value = df['Graduated'].mode()[0] 
df['Graduated'].fillna(mode_value, inplace = True)
df
```

```python
median_value = df['Age'].median()
df['Age'].fillna(median_value, inplace = True)
df
```

```python
df['Enrollment_Date'] = pd.to_datetime(df['Enrollment_Date'], errors='coerce')
df
```

```python
mean_date = df['Enrollment_Date'].mean()
mean_date = mean_date.normalize() if pd.notna(mean_date) else pd.NaT
df['Enrollment_Date'].fillna(mean_date, inplace =True)
df
```

```python
mean_value = df['Math_Score'].mean()
df['Math_Score'].fillna(mean_value, inplace = True)
df['Science_Score'].fillna(mean_value, inplace = True)
df
```

```python
df['Age'] = df['Age'].astype(int)
df['Math_Score'] = pd.to_numeric(df['Math_Score'], errors='coerce') 
df['Science_Score'] = pd.to_numeric(df['Science_Score'], errors='coerce')
df.dtypes
```

```python
df['Math_Score'] = df['Math_Score'].clip(80,100)
```

```python
df['Math_Score']
```

```python
lower_bound = df['Science_Score'].quantile(0.25)
upper_bound = df['Science_Score'].quantile(0.75)
df['Science_Score'] = df['Science_Score'].clip(lower=lower_bound, upper=upper_bound)
df
```

```python
data = {
    'Math Score': [88, 92, 100, 80, 85, 89.3, 100, 95, 80, 88, 89.3, 85],
    'Science Score': [89.3, 85, 93.29, 76.104, 89.3, 93.29, 93.29, 93.29, 76.104, 77, 85, 76.104],
    'Graduated': ['yes', 'no', 'no', 'yes', 'no', 'no', 'yes', 'no', 'yes', 'no', 'no', 'yes']
}

df = pd.DataFrame(data)
ig, axes = plt.subplots(1, 2, figsize=(15, 7), sharey=True)

sns.violinplot(x='Graduated', y='Math Score', data=df, ax=axes[0])
axes[0].set_title('Distribution of Math Scores by Graduation Status')

sns.violinplot(x='Graduated', y='Science Score', data=df, ax=axes[1])
axes[1].set_title('Distribution of Science Scores by Graduation Status')
plt.tight_layout()
plt.show()
```

```python
df_melted = df.melt(id_vars='Graduated', var_name='Subject', value_name='Score')
plt.figure(figsize=(12, 6))
sns.boxplot(x='Graduated', y='Score', hue='Subject', data=df_melted)
plt.title('Box Plot of Scores by Graduation Status')
plt.show()
```

```python
math_scores = [88, 92, 100, 80, 85, 89.3, 100, 95, 80, 88, 89.3, 85]
science_scores = [89.3, 85, 93.29, 76.104, 89.3, 93.29, 93.29, 93.29, 76.104, 77, 85, 76.104]
df = pd.DataFrame({
    'Math': math_scores,
    'Science': science_scores
})
plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
sns.kdeplot(df['Math'], fill=True, color='blue')
plt.title('KDE Plot of Math Scores')
plt.xlabel('Score')
plt.ylabel('Density')
plt.subplot(1, 2, 2)
sns.kdeplot(df['Science'], fill=True, color='green')
plt.title('KDE Plot of Science Scores')
plt.xlabel('Score')
plt.ylabel('Density')
plt.tight_layout()
plt.show()
```
